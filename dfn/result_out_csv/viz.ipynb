```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# sns.set()
# sns color palette
# sns.set_palette('colorblind')
sns.set_palette('deep')
# sns background grid
sns.set_style('whitegrid')
# Load the CSV files
df1 = pd.read_csv('cos_in_outs.csv')
df2 = pd.read_csv('exp_sin_in_outs.csv')
df3 = pd.read_csv('sin_in_outs.csv')
# Plot the data in subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10,4.5))

# Plot for df1
x = np.linspace(0, 2*np.pi, 100)
y = np.cos(x)
ax1.plot(df1['input'], df1['output'], label='ADFN', linewidth=4, color='r')
ax1.plot(x, y, label='cos(x)', linewidth=3)
ax1.set_xlabel('x', fontsize=16)
ax1.set_ylabel('y', fontsize=16)
ax1.legend(fontsize=16)
ax1.set_title('(a)', fontsize=16)

# Plot for df2
x = np.linspace(-np.pi, np.pi, 100)
y = np.exp(x) + np.sin(x)
ax2.plot(df2['input'], df2['output'], label='ADFN', linewidth=4, color='r')
ax2.plot(x, y, label='exp(x) + sin(x)', linewidth=3)
ax2.set_xlabel('x', fontsize=16)
# ax2.set_ylabel('Y', fontsize=16)
ax2.legend(fontsize=16)
ax2.set_title('(b)', fontsize=16)

# Plot for df3
x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)
ax3.plot(df3['input'], df3['output'], label='ADFN', linewidth=4, color='r')
ax3.plot(x, y, label='sin(x)', linewidth=3)
ax3.set_xlabel('x', fontsize=16)
# ax3.set_ylabel('y', fontsize=16)
ax3.legend(fontsize=16)
ax3.set_title('(c)', fontsize=16)

plt.tight_layout()
plt.savefig('mixed_functions.pdf')
```
